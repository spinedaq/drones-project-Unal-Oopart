#Turning off the classes to the validation set anotations:

```python
names_txt=os.listdir("/content/labels/val/content/drive/MyDrive/Yolo_v6/datasets_usados_estructurados/VisDrone2019-VID-val/labels/val") #Names of the txt files of the validation set
#traverse the txt files of the respective set:
for j in names_txt:
  txt=pd.read_csv("/content/labels/val/content/drive/MyDrive/Yolo_v6/datasets_usados_estructurados/VisDrone2019-VID-val/labels/val/"+j, sep=" ", header=None) #Read the txt file in a DataFrame
  txt=txt.to_numpy() #Go the Dataframe to a numpy array
  positions_0=np.where(txt[...,0]==0) #Positions of the rows where there are class number 0(ignored regions)
  txt=np.delete(txt,list(positions_0[0]),axis=0) #Delete the rows where there are class number 0(ignored regions)
  positions_11=np.where(txt[...,0]==11) #Positions of the rows where there are class number 11(others)
  txt=np.delete(txt,list(positions_11[0]),axis=0) #Delete the rows where there are class number 11(others)
  #traverse the rows of the numpy array for subtract 1 to every class because the class stayed a unity more ahead. 
  for k in range(txt.shape[0]):
    txt[k,0]=txt[k,0]-1
  txt=pd.DataFrame(txt) #Go the numpy array to a DataFrame
  txt[0]=txt[0].astype(int) #The class should be int
  #write the information in the txt file again
  file = open("/content/labels/val/content/drive/MyDrive/Yolo_v6/datasets_usados_estructurados/VisDrone2019-VID-val/labels/val/"+j, "w") #Donde guardaré el archvio y que sea para escritura
  index=txt.columns.values #Nombres de las columnas del dataframe
  for l in range(len(txt.index.values)):
    for m in index:
      if m!=index[len(index)-1]:
        file.write(str(txt[m][l])) #Escribiendo columna por columna en el txt
        file.write(" ") #Entre cada columna hay un espacio
      else:
        file.write(str(txt[m][l])+os.linesep) #Para el salto de linea
  file.close() #Cierro el file txt ya terminado
```
